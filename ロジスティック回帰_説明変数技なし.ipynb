```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report

data = pd.read_csv('kairyu-2.csv')  # ヘッダーがないことを指定
data
```

```python
data.info()
```

```python
print(data.isnull().values.sum())
```

```python
df = data.dropna()
df
```

```python
df.info()
```

```python
df.columns
```

```python
df = pd.get_dummies(df,columns=['性格', '特性', 'テラスタイプ'])
df
```

```python
df.info()
```

```python
df['持ち物'].unique()
```

```python
df['持ち物'] = df['持ち物'].replace({
    'あおぞらプレート': 1,
    'あつぞこブーツ': 2,
    'いかさまダイス': 3,
    'おんみつマント': 4,
    'クリアチャーム': 5,
    'こだわりハチマキ': 6,
    'ゴツゴツメット': 7,
    'じゃくてんほけん': 8,
    'するどいくちばし': 9,
    'たべのこし': 10,
    'タラプのみ': 11,
    'とつげきチョッキ': 12,
    'ラムのみ': 13
})
df.info()
```

```python
df_train = df[:180]
df_train
```

```python
df_test = df[180:]
df_test
```

```python
y = df_train["持ち物"]
X = df_train.drop(columns=["技1", "技2", "技3", "技4", "持ち物"])
y
```

```python
#yがint型ではなかったらやる
y = y.astype(int)
y
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
import numpy as np


X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)#150までのデー

# print(type(X_train))
# print(type(y_train))

# display(y_train)
# y_train = np.array(y_train)
# print(type(y_train))
# ロジスティック回帰モデルの作成と訓練
model = LogisticRegression()
model.fit(X_train, y_train)

# モデルの評価
predictions = model.predict(X_test)
print(classification_report(y_test, predictions))
```

```python
y_ev = df_test["持ち物"]
X_ev = df_test.drop(columns=["技1", "技2", "技3", "技4", "持ち物"])
y_ev
```

```python
#y_evがint型ではなかったらやる
y_ev = y_ev.astype(int)
y_ev
```
